### **Employee Attrition is Highly Related to the Job Roles of Sales Executive,Sales Representative, Laboratory Technician, and Research Scientist in the Sales and Research & Development Departments**

```python
# Relation between Total Working Years and the Working Years at the Company and the Years Since Last Promotion of Former Employees
plt.figure(figsize=(10, 5))
sns.scatterplot(x='TotalWorkingYears', y='YearsAtCompany', hue='YearsSinceLastPromotion',data=df3, palette='YlGnBu')
plt.title('Total Working Years Vs Working Years at the Company')
plt.xlabel('Total Working Years')
plt.ylabel('Working Years at the Company')
plt.legend(title='Years Since Last Promotion')
plt.show()
```

### **The Tendency of Early-Career Employees to Leave is Higher**

```python
# df4 => The new Dataframe includes only former employees working in the Sales and Research & development Department
specific_JobRole=['Laboratory Technician','Research Scientist','Sales Executive','Sales Representative']
df4=df[df['JobRole'].isin(specific_JobRole)]
df4=df4.loc[~df4['Department'].str.contains('Human Resources')]
```

```python
# New Columns that Combines the Job Role and Job Level
df4['Job Role & Level']=df4['JobRole']+ '_' + df4['JobLevel'].astype(str)
```

```python
# df5 => New DataFrame to extract a table of the average monthly income of the Job Role & Level and the Employee Attrition from Dataframe df4
df5=df4.groupby(['Job Role & Level','Attrition']).agg(mean_monthly_income=('MonthlyIncome', 'mean'))
df5 = df5.reset_index()
df5.head()
```

```python
# Bar Plot of the average monthly income of each Job Role & Level in terms of Employee Attrition
plt.figure(figsize=(25,5))
a=sns.barplot(x="Job Role & Level",
           y="mean_monthly_income",
           hue="Attrition",
           data=df5)
for label in a.containers:
  a.bar_label(label)
plt.title('Average Monthly Income by Job Role & Level of Employee Attrition');
plt.ylabel('Average Monthly Income')
```

### **Employees Who Earn More or Less than the Average for their Job Role and Level are Leaving the Company**
